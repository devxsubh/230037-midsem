# Task 1.3 — What the Paper Claims to Improve

**Paper:** Logical-Shapelets (Mueen, Keogh, Young, KDD 2011)

---

## Main baseline or prior method

The main baseline the paper compares against is the **original shapelet method** (Ye & Keogh, "Time Series Shapelets: A New Primitive for Data Mining," KDD 2009)—i.e. classic shapelets discovered by the brute-force algorithm (Algorithm 1 in the 2011 paper, which mirrors the original discovery process) and used for binary splits based on a single shapelet and threshold.

---

## Limitation of the baseline the paper identifies

The paper identifies two limitations: (1) **Expressiveness**—classic shapelets are limited to binary presence/absence questions ("is this pattern close or not?"). They cannot express concepts that require, for example, the co-occurrence of two patterns (AND) or the presence of one of several patterns (OR). The paper gives the example of a rectangular gesture that has variable pause times at corners: one shapelet cannot describe the class; a conjunction of shapelets (e.g. α and β and γ and δ) can. (2) **Speed**—the original algorithm is slow (large number of candidates and expensive distance computations), so it does not scale well to long or large datasets.

---

## How the proposed method overcomes this limitation

The paper introduces **logical shapelets** (Section 4): combinations of shapelets with AND and OR. For AND, a time series is assigned to the "close" side only if it is close to *all* of the constituent shapelets (combined distance = maximum of the individual distances); for OR, if it is close to *any* (combined distance = minimum). This increases expressiveness. The paper also introduces **two speedup techniques** (Section 3): sufficient statistics for O(1) distance computation per alignment (Section 3.1) and admissible pruning to skip candidates that cannot beat the current best (Section 3.2), so that both classic and logical shapelets can be found much faster.

---

## One scenario where the paper's method would not outperform the baseline

When the classes are already **perfectly or very well separated by a single shapelet** and the dataset has little noise, the added expressiveness of logical shapelets does not help. In that case, the classic shapelet (one pattern, one threshold) is sufficient, and the extra search for a second shapelet (and the risk of overfitting a more complex logical rule) could even slightly hurt accuracy or interpretability. The paper itself notes that logical shapelets "degenerate to classic shapelets" when the data does not need the extra expressiveness (Section 1), so in such a scenario the method would not outperform—and might match or marginally underperform—the baseline in terms of simplicity and generalization.
