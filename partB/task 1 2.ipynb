# Task 1.2 — Key Assumptions

**Paper:** Logical-Shapelets (Mueen, Keogh, Young, KDD 2011)

---

## Assumption 1

**Assumption:** Time series (and subsequences) are z-normalized before distance computation, so that scale and offset differences do not dominate; the method assumes that *shape* (relative variation over time) is discriminative rather than absolute scale or baseline.

**Why the method needs it:** The subsequence distance (Eq. 2, Section 2) and the quality measure (information gain) are defined on normalized distances. Without z-normalization, two subsequences with the same shape but different amplitudes would be considered far apart, and the shapelet would fail to match the same pattern across instances.

**Violation scenario:** In a dataset where the *magnitude* of a peak (e.g. amplitude of a heartbeat) is the discriminative feature and the shape is similar across classes, z-normalization would remove that signal and the method would struggle to separate classes.

**Paper reference:** Section 2 ("To achieve scale and offset invariance, we must normalize the individual time series using z-normalization before the actual distance is computed").

---

## Assumption 2

**Assumption:** The discriminative information can be captured by a *single* local subsequence (or a small logical combination of such subsequences) per split—i.e. the method assumes that the class boundary can be expressed as "close to pattern A (and/or pattern B)" versus "far."

**Why the method needs it:** The entire representation is built on shapelets (subsequences) and thresholds. If the true concept requires counting occurrences, or depends on global statistics (e.g. mean over the whole series), or on patterns that are not contiguous subsequences, the shapelet model cannot represent it.

**Violation scenario:** A dataset where the two classes differ only in the *number* of times a certain motif appears (e.g. class 1 has the motif twice, class 2 once). A single shapelet asks only presence/absence; the paper notes this limitation and uses logical shapelets to partly address it, but complex counting concepts may still be out of scope.

**Paper reference:** Section 1 ("differentiating classes by using only binary questions is not possible if the classes differ only in the number of occurrences") and Section 4 (logical shapelets as an extension).

---

## Assumption 3

**Assumption:** The quality of a shapelet is appropriately measured by *information gain* of the split induced by a distance threshold (with tie-breaking by separation gap). The method assumes that maximizing information gain leads to good generalization.

**Why the method needs it:** The search algorithm (Algorithm 1, 3) evaluates every candidate by its best possible information gain. If information gain were not a suitable criterion for this problem (e.g. if the classes are not separable by a threshold on distance), the chosen shapelet would be poor and classification accuracy would suffer.

**Violation scenario:** When the optimal decision boundary in the distance space is not a single threshold (e.g. one class is "close" for small distances and also for large distances, with the other class in between). In such a case, a single threshold cannot achieve high information gain and the method would underperform.

**Paper reference:** Definition 3 (information gain), Definition 5 (shapelet maximizes information gain), Algorithm 3 (bestIG).
