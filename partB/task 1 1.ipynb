# Task 1.1 — Core Contribution / Architecture

**Paper:** Logical-Shapelets: An Expressive Primitive for Time Series Classification (Mueen, Keogh, Young, KDD 2011)

## Step-by-Step Method Description

**Step 1: Candidate generation**
- The algorithm enumerates all possible subsequences from the training time series: for each series, for each length (from 1 to minimum series length), and for each starting position. Each such subsequence is a shapelet candidate.
- *Reference:* Algorithm 1, lines 3–6 (loops over j, l, i).
- *Purpose:* To search over the space of local patterns that could discriminate between classes.

**Step 2: Subsequence distance computation**
- For each candidate subsequence, the algorithm computes its distance to every time series in the dataset. The distance is the *subsequence distance*: the minimum over all alignments (sliding the candidate along the series) of the z-normalized Euclidean distance between the candidate and the aligned window.
- *Reference:* Section 2 (Definition of sdist), Eq. 2 (normalized Euclidean distance), Algorithm 2.
- *Purpose:* Z-normalization gives scale and offset invariance; the minimum over alignments finds the best match of the pattern within each series.

**Step 3: Orderline and split evaluation**
- The distances of all series to the candidate form an "orderline": each series is placed on a line according to its distance to the candidate. The algorithm then evaluates every possible split threshold (midpoints between consecutive distances) and computes the information gain of splitting the data into "close" (distance ≤ τ) and "far" (distance > τ).
- *Reference:* Figure 2 (orderline), Algorithm 3 (bestIG), Definition 3 (information gain).
- *Purpose:* To find the best threshold τ that separates classes when using this candidate as a shapelet.

**Step 4: Best shapelet selection**
- The candidate that achieves the maximum information gain (with ties broken by separation gap, Definition 4) is stored as the best shapelet so far. The process repeats until all candidates are evaluated.
- *Reference:* Algorithm 1 lines 10–12, Definition 5 (shapelet).
- *Purpose:* To select the single discriminative pattern that best splits the dataset.

**Step 5 (optional): Logical combination**
- The paper extends to logical shapelets: instead of one shapelet, two (or more) shapelets can be combined with AND (both must be close) or OR (at least one close). For AND, the "distance" of a series is the maximum of its distances to the two shapelets; for OR, the minimum. Information gain is then computed on this combined orderline.
- *Reference:* Section 4 (Logical Shapelet), Figure 5.
- *Purpose:* To represent concepts that require the co-occurrence (AND) or alternative presence (OR) of multiple patterns.

**Step 6: Output**
- The final output is the best shapelet (or pair of shapelets with a logical operator) together with the optimal threshold τ, which can be used to classify new time series by comparing their subsequence distance(s) to the threshold(s).

## Final Summary Sentence

The paper addresses the limitations of classic shapelets—limited expressiveness (binary presence/absence only) and slow discovery—by introducing logical combinations (AND/OR) of shapelets and two algorithmic speedups (sufficient statistics for O(1) distance per alignment and admissible pruning). The authors claim that logical shapelets are more expressive and that their faster algorithm makes discovery practical on benchmark datasets.
